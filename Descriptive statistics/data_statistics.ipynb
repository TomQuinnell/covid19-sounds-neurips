## User numbers

```python
df = pd.read_csv("results_raw_20210426_lan_yamnet_android_noloc.csv",sep = ';')
df = df.drop(['Date','Folder Name','Symptoms','Covid-Tested','Hospitalized',
              'Voice filename','Cough filename','Breath filename','Voice check','Cough check','Breath check'],axis=1)
df_android = df.drop_duplicates()

df = pd.read_csv("results_raw_20210426_lan_yamnet_ios_noloc.csv",sep = ';')
df = df.drop(['Date','Folder Name','Symptoms','Covid-Tested','Hospitalized',
              'Voice filename','Cough filename','Breath filename','Voice check','Cough check','Breath check'],axis=1)
df_ios = df.drop_duplicates()

df = pd.read_csv("results_raw_20210426_lan_yamnet_web_noloc.csv",sep = ';')
df = df.drop(['Uid','Folder Name','Symptoms','Covid-Tested','Hospitalized',
              'Voice filename','Cough filename','Breath filename','Voice check','Cough check','Breath check'],axis=1)
df_web = df.drop_duplicates()
df_web = df_web.rename(columns={'Date':'Uid'})
```
